```python
#Quiz options in user menu

#Function to read quiz 1 from txt file, will upgrade this to database once program logic is more fleshed out
def take_quiz1():
    with open("quiz1.txt","r") as f:    #Open file and declare variable
        line_count = 0                  #Used to count number of questions asked
        correct = 0                     #Used to count number of correct answers
        incorrect = 0                   #Number of incorrect answers. Not currently in use but may be useful later?
        file = f.readlines()            #Readlines method to go line by line
        for line in file:
            line_list = file[line_count].strip().split(",")    #Seperate each data point by comma
            line_count+=1
            print (line_list[0])            #Prints question number
            print (line_list[1])            #Prints question
            print ("a. " + line_list[2])    #Prints option a
            print ("b. " + line_list[3])
            print ("c. " + line_list[4])
            choice = " " + input("")        #Takes user input
            if choice == line_list[5]:      #Checks against the last item on the line (where correct answer is listed)
                correct += 1                #If correct, add to total
                print("yes\n")
            else:
                incorrect += 1
                print("\n")

        print ("\n")
        print ("correct: {}/{}".format(correct, line_count))    #Display results
        

#Copy the exact same logic but for different file (or database in the future)
def take_quiz2():
    with open("quiz2.txt","r") as f:
        line_count = 0
        correct = 0
        incorrect = 0
        file = f.readlines()
        for line in file:
            line_list = file[line_count].strip().split(",")
            line_count+=1
            print (line_list[0]) 
            print (line_list[1])
            print ("a. " + line_list[2])
            print ("b. " + line_list[3])
            print ("c. " + line_list[4])
            choice = " " + input("")
            if choice == line_list[5]:
                correct += 1
                print("yes\n")
            else:
                incorrect += 1
                print("\n")

        print ("\n")
        print ("correct: {}/{}".format(correct, line_count))
        

def take_quiz3():
    with open("quiz3.txt","r") as f:
        line_count = 0
        correct = 0
        incorrect = 0
        file = f.readlines()
        for line in file:
            line_list = file[line_count].strip().split(",")
            line_count+=1
            print (line_list[0]) 
            print (line_list[1])
            print ("a. " + line_list[2])
            print ("b. " + line_list[3])
            print ("c. " + line_list[4])
            choice = " " + input("")
            if choice == line_list[5]:
                correct += 1
                print("yes\n")
            else:
                incorrect += 1
                print("\n")

        print ("\n")
        print ("correct: {}/{}".format(correct, line_count))
        
        
#-------------------------------------------------------------------------------------------------------------------#

        
#Function to choose which quiz to take
def quiz_choice():
    
    #While loop to keep menu running after quiz, or in case of invalid input
    while True:
        quiz_select = input(" (a) quiz 1\n (b) quiz 2\n (c) Go Back\n")
        if quiz_select[0].lower() == "a":    #Use .lower() in case user enters a higher case letter
            take_quiz1()                     #Call method to take relevant quiz
            print("Go Again?")
        elif quiz_select[0].lower() == "b":
            take_quiz2()                     #Call method to take relevant quiz
            print("Go Again?")
        elif quiz_select[0].lower() == "c":   #Goes back one step in the menu 
            break
        else:
            continue
            

#---------------------------------------------------------------------------------------------------------------------#
        

#Function to allow user to choose a quiz or go back to main to be quiz master
def user_menu():
    
    #Again while loop to keep menu running
    while True:
        
        user_choice = input(" (a) Choose Quiz\n (b) Go Back\n")
        if user_choice[0].lower() == "b":
            break
        elif user_choice[0].lower() == "a":
            quiz_choice()
            
            
#---------------------------------------------------------------------------------------------------------------------#


#Main menu where user can choose between taking quizzes or being quiz master
def main_menu():
    while True:

        menu_choice = input(" a) Take Quiz\n b) Quiz Master\n c) Quit\n")
        if menu_choice[0].lower() == "a":
            user_menu()
        elif menu_choice[0].lower() == "b":
            master_menu()
        elif menu_choice[0].lower() == "c":
            print("Bye:)")
            break     
            
            
#--------------------------------------------------------------------------------------------------------------------#


#Function to add quiz
def add_quiz():
    question_count = 0   #question count adds line number to txt file
    num_of_questions = 3    #Set to 3 questions for for loop
    quiz_title = input("Name of quiz: ")      #Get desired quiz name from user input
    for i in range(num_of_questions):
        question = input("Type Question")     #Take question
        option1 = input("Give option a")      #User gives three possible answer
        option2 = input("Give option b")
        option3 = input("Give option c")
        answer = input("Which was the correct one?")   #User to indicate letter of correct answer
        
        #Create string / line to be written to file and declare it as variable
        write_to_file = (f"{question_count+1}, {question}, {option1}, {option2}, {option3}, {answer}")  
        
        #Creates new file with quiz title chosen by user earlier
        with open (f"{quiz_title}.txt","a") as new_quiz:
            new_quiz.write(write_to_file+"\n")            #add write_to_file variable to txt file
            print("Done")
    
    #Call notify() method to send email to devs/support to add new option to quiz menu
    notify()


#--------------------------------------------------------------------------------------------------------------------#


import smtplib

#Function to notify support engineers to load newly created quiz
#Resource: https://www.youtube.com/watch?v=Y_tnWTjTfzY
def notify():
    server = smtplib.SMTP_SSL("smtp.gmail.com", 465)                        #Connect to gmail server
    server.login("wezquizdevs@gmail.com", "uslmrvunjrhlvkob")               #Login to sender email address
    server.sendmail("wezquizdevs@gmail.com",                                
                    "jayrbsn@gmail.com",                                    #Sends to your own email to test funtion                                     
                    "New Quiz Notification. Please add menu option")

    server.quit()
    print ("Quiz Added. The quiz offering will be updated within 48hrs")
    

#--------------------------------------------------------------------------------------------------------------------#


#Note to self: change options to letters
#Function to run menu for quiz masters
def master_menu():
    
    while True:
        try:
            menu_choice = int(input('''Please choose an option from the menu:\n
                  1.View Quizzes 
                  2.Create Quiz 
                  3.Edit Quiz 
                  4.Report Quiz 
                  5.Back\n: '''))
       
            if menu_choice == 1:
                quiz_counter()                    #Call quiz_counter() method to print out quiz list
                
                
            elif menu_choice == 2:
                add_quiz()                        #Calls add_quiz() to create new quiz + notify devs to add to program
                

            elif menu_choice == 3:
                print("Edit option coming soon")
                continue

            elif menu_choice == 4:
                print("Reporting option coming soon")
                continue

            elif menu_choice == 5:
                break                              #Goes back to main menu
                
            else:
                print("Oops. Please type a number from 1-5 and hit enter")
                continue
        except:
            print("you suck ass")    #For testing only. Will change to more professional before final submission
            
            
#---------------------------------------------------------------------------------------------------------------------#


import os

#Function to display existing quizzes to quiz masters
def quiz_counter():
    x=0
    quiz_list = []                                              #Create list to store quiz names 
    path = (r"C:\Users\Jarryd\Desktop\HyperionDev\WezQuiz")     #File path will obviously depend on circumstance
                                                                
    files = os.listdir(path)

    for f in files:
        if f.endswith('.txt') and f not in quiz_list:
            f_name = os.path.splitext(f)[0]                     #removes .txt part of file name
            quiz_list.append(f_name)                            #Adds file to list of quizzes
            

    for item in quiz_list:                               #For loop to iterate through list of quizzes and print them out
        x+=1
        #item.splitext(",")
        print ("{}. {}".format(x, item))
        
    
    #choose = input("enter:")
    
    print("\n")
    #master_menu()

```
